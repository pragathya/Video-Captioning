```python
# Import necessary libraries
import os
import shutil
import json
import traceback

# Function to load data from video files and corresponding captions
def load_data(video_folder, caption_file, num_videos=100):
    # Get a list of video file names in the specified folder
    video_files = [file.strip() for file in os.listdir(video_folder) if file.endswith(".avi")]
    video_files = [file.replace(".avi", "") for file in video_files]
    video_files = video_files[:num_videos]

    # Print the list of video files (for debugging purposes)
    print(video_files)

    # Read captions from the specified file
    with open(caption_file, 'r') as f:
        captions = f.readlines()

    # Create a mapping between video IDs and their corresponding captions
    video_caption_mapping = {}
    for caption in captions:
        try:
            # Split each line into video ID and caption text
            line = caption.strip().split(' ', 1)
            video_id = line[0]
            caption_text = line[1]

            # Check if the video ID is in the selected list of video files
            if video_id in video_files:
                # If the video ID is not in the mapping, create a new entry
                if video_id not in video_caption_mapping:
                    video_caption_mapping[video_id] = []
                # Append the caption text to the corresponding video ID
                video_caption_mapping[video_id].append(caption_text)
        except Exception as e:
            # Print any exceptions that occur during processing
            print(e)

    # Return the final mapping of video IDs to captions
    return video_caption_mapping

# Set the paths for video files, captions, and output folder
video_folder = "/content/drive/MyDrive/VideoCaptioning/dataset/YouTubeClips"
caption_file = "/content/drive/MyDrive/VideoCaptioning/dataset/MSVD/AllVideoDescriptions.txt"
output_folder = "/content/drive/MyDrive/VideoCaptioning/data/training_data/video"
json_file_path = "/content/drive/MyDrive/VideoCaptioning/data/training_data/training_list.json"

# Load data for the first 200 videos
video_caption_mapping = load_data(video_folder, caption_file, num_videos=500)
# Print the mapping of video IDs to captions (for debugging purposes)
print(video_caption_mapping)

# Create a list to store the data in the required format
training_list = []

# Copy the first 500 videos to the output folder
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for video_file in list(video_caption_mapping.keys())[:500]:
    # Set the source and destination paths for video files
    source_path = os.path.join(video_folder, video_file+".avi")
    destination_path = os.path.join(output_folder, video_file+".avi")

    # Check if the file already exists in the output folder
    if os.path.exists(destination_path):
        print(f"Skipping video {video_file}, file already exists.")
    else:
        # Copy the video file to the output folder
        shutil.copyfile(source_path, destination_path)

    # Create a JSON entry for each video with its captions and ID
    json_entry = {
        'caption': video_caption_mapping[video_file],
        'id': video_file.split('.')[0]  # Remove the file extension
    }
    training_list.append(json_entry)

# Print the final list of JSON entries (for debugging purposes)
print(training_list)

# Write the JSON file containing the training data
with open(json_file_path, 'w') as json_file:
    json.dump(training_list, json_file, indent=2)

# Print a success message
print(f"Training list and video files copied successfully to {output_folder}.")
```

```python
import shutil
import tqdm
import numpy as np
import cv2
import os
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.models import Model
from concurrent.futures import ProcessPoolExecutor

# Function to convert a video into frames and return a list of image paths
def video_to_frames(video):
    # Create a temporary directory for storing video frames
    path = os.path.join(config.test_path, 'temporary_images_'+video)
    if os.path.exists(path):
        shutil.rmtree(path)
    os.makedirs(path)
    video_path = "/content/drive/MyDrive/VideoCaptioning/data/training_data/video/" + video
    count = 0
    image_list = []

    # Open the video file
    cap = cv2.VideoCapture(video_path)
    frame_skip = 10  # Number of frames to skip

    while cap.isOpened():
        ret, frame = cap.read()
        if ret is False:
            break

        # Save frames at specified intervals
        if count % (frame_skip + 1) == 0:
            cv2.imwrite(os.path.join(config.test_path, 'temporary_images_'+video, 'frame%d.jpg' % count), frame)
            image_list.append(os.path.join(config.test_path, 'temporary_images_'+video, 'frame%d.jpg' % count))

        count += 1

    cap.release()
    cv2.destroyAllWindows()
    return image_list

# Function to load the VGG16 model for feature extraction
def model_cnn_load():
    # Load the VGG16 model with pre-trained weights
    model = VGG16(weights="imagenet", include_top=True, input_shape=(224, 224, 3))
    out = model.layers[-2].output
    model_final = Model(inputs=model.input, outputs=out)
    return model_final

# Function to load an image from a given path and resize it
def load_image(path):
    # Read and resize an image
    img = cv2.imread(path)
    img = cv2.resize(img, (224, 224))
    return img

# Function to extract features from a video using the provided model
def extract_features(video, model):
    # Get video ID and print processing message
    video_id = video.split(".")[0]
    print(video_id)
    print(f'Processing video {video}')

    # Get a list of frames from the video
    image_list = video_to_frames(video)

    # Sample frames for feature extraction
    samples = np.round(np.linspace(0, len(image_list) - 1, 80))
    image_list = [image_list[int(sample)] for sample in samples]

    # Initialize an array for storing images
    images = np.zeros((len(image_list), 224, 224, 3))

    # Load and store images in the array
    for i in range(len(image_list)):
        img = load_image(image_list[i])
        images[i] = img

    images = np.array(images)

    # Extract features using the model
    fc_feats = model.predict(images, batch_size=128)
    img_feats = np.array(fc_feats)

    # Cleanup temporary image directory
    shutil.rmtree(os.path.join(config.test_path, 'temporary_images_'+video))
    return img_feats

# Function to process a video, extract features, and save them to a file
def process_video(video, model):
    # Define the output file path
    outfile = '/content/drive/MyDrive/VideoCaptioning/data/training_data/feat/' + video + '.npy'

    # Check if the file already exists
    if os.path.exists(outfile):
        print(f"Skipping video {video}, file already exists.")
    else:
        # Extract features and save them to a file
        img_feats = extract_features(video, model)
        np.save(outfile, img_feats)
        print(f"Features saved for video {video}.")

# Function to extract features for all videos using multithreading
def extract_feats_pretrained_cnn():
    # Load the VGG16 model
    model = model_cnn_load()
    print('Model loaded')

    # Create the output directory if it does not exist
    if not os.path.isdir("/content/drive/MyDrive/VideoCaptioning/data/training_data/feat/"):
        os.mkdir("/content/drive/MyDrive/VideoCaptioning/data/training_data/feat/")

    # Get a list of video files
    video_list = os.listdir("/content/drive/MyDrive/VideoCaptioning/data/training_data/video")

    # Remove '.ipynb_checkpoints' if it exists
    try:
        video_list.remove('.ipynb_checkpoints')
    except:
        pass

    # Use simple threading
    import threading
    import time

    # Use simple threading
    threads = []
    count = 0

    # Process each video in a separate thread
    for video in video_list:
        thread = threading.Thread(target=process_video, args=(video, model))
        threads.append(thread)
        thread.start()

        count += 1
        if count % 10 == 0:
            print(f"Processed {count} videos. Waiting for 1 minute...")
            time.sleep(120)  # Sleep for 1 minute

    # Wait for all threads to finish
    for thread in threads:
        thread.join()

import concurrent.futures

# Entry point of the script
if __name__ == "__main__":
    extract_feats_pretrained_cnn()
```

```python
import json
import os
import random

import keras
import numpy as np
from keras.callbacks import EarlyStopping
from keras.layers import Input, LSTM, Dense
from keras.models import Model
from keras.preprocessing.sequence import pad_sequences
from keras.preprocessing.text import Tokenizer
from keras.utils import to_categorical
import joblib
```

```python
from keras.callbacks import ModelCheckpoint
from keras.models import Model
from keras.layers import Input, LSTM, Dense
from keras.utils import to_categorical
from keras.callbacks import EarlyStopping
import os
import random
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import json
import numpy as np
from keras.optimizers import Adam
from sklearn.externals import joblib  # Import joblib for saving Tokenizer

# Class definition for training video description model
class VideoDescriptionTrain(object):
    """
    Initialize the parameters for the model
    """

    def __init__(self, config):
        # Configuration parameters
        self.train_path = config.train_path
        self.test_path = config.test_path
        self.max_length = config.max_length
        self.batch_size = config.batch_size
        self.lr = config.learning_rate
        self.epochs = config.epochs
        self.latent_dim = config.latent_dim
        self.validation_split = config.validation_split
        self.num_encoder_tokens = config.num_encoder_tokens
        self.num_decoder_tokens = config.num_decoder_tokens
        self.time_steps_encoder = config.time_steps_encoder
        self.time_steps_decoder = 80
        config.time_steps_decoder = 80
        self.x_data = {}

        # Processed data
        self.tokenizer = None

        # Models
        self.encoder_model = None
        self.decoder_model = None
        self.inf_encoder_model = None
        self.inf_decoder_model = None
        self.save_model_path = config.save_model_path

    def preprocessing(self):
        """
        Preprocessing the data
        Dumps values of the json file into a list
        """
        TRAIN_LABEL_PATH = '/content/drive/MyDrive/VideoCaptioning/data/training_data/training_list.json'
        with open(TRAIN_LABEL_PATH) as data_file:
            y_data = json.load(data_file)
        train_list = []
        vocab_list = []
        for y in y_data:
            for caption in y['caption']:
                caption = "<bos> " + caption + " <eos>"
                if len(caption.split()) > 10 or len(caption.split()) < 6:
                    continue
                else:
                    train_list.append([caption, y['id']])

        random.shuffle(train_list)
        training_list = train_list[int(len(train_list) * self.validation_split):]
        validation_list = train_list[:int(len(train_list) * self.validation_split)]
        for train in training_list:
            vocab_list.append(train[0])
        self.tokenizer = Tokenizer(num_words=self.num_decoder_tokens)
        self.tokenizer.fit_on_texts(vocab_list)

        TRAIN_FEATURE_DIR = "/content/drive/MyDrive/VideoCaptioning/data/training_data/feat"
        for filename in os.listdir(TRAIN_FEATURE_DIR):
            f = np.load(os.path.join(TRAIN_FEATURE_DIR, filename), allow_pickle=True)
            self.x_data[filename[:-4]] = f
        return training_list, validation_list

    def load_dataset(self, training_list):
        """
        Loads the dataset in batches for training
        :return: batch of data
        """
        encoder_input_data = []
        decoder_input_data = []
        decoder_target_data = []
        videoId = []
        videoSeq = []
        for idx, cap in enumerate(training_list):
            caption = cap[0]
            videoId.append(cap[1])
            videoSeq.append(caption)
        train_sequences = self.tokenizer.texts_to_sequences(videoSeq)
        train_sequences = np.array(train_sequences)
        train_sequences = pad_sequences(train_sequences, padding='post', truncating='post',
                                        maxlen=self.max_length)
        file_size = len(train_sequences)
        n = 0

        for i in range(self.epochs):

            for idx in range(0, file_size):

                if videoId[idx] + ".avi" not in self.x_data:
                    pass  # Warning: Video ID not found in self.x_data.
                else:
                    n += 1
                    encoder_input_data.append(self.x_data[videoId[idx] + ".avi"])
                    y = to_categorical(train_sequences[idx], self.num_decoder_tokens)
                    decoder_input_data.append(y[:-1])
                    decoder_target_data.append(y[1:])
                    if n == self.batch_size:
                        encoder_input = np.array(encoder_input_data)
                        decoder_input = np.array(decoder_input_data)
                        decoder_target = np.array(decoder_target_data)
                        encoder_input_data = []
                        decoder_input_data = []
                        decoder_target_data = []
                        n = 0
                        yield ([encoder_input, decoder_input], decoder_target)

    def train_model(self):
        # Encoder model
        encoder_inputs = Input(shape=(config.time_steps_encoder, config.num_encoder_tokens), name="encoder_inputs")
        encoder = LSTM(config.latent_dim, return_state=True, return_sequences=True, name='encoder_lstm')
        _, state_h, state_c = encoder(encoder_inputs)
        encoder_states = [state_h, state_c]

        # Decoder model
        decoder_inputs = Input(shape=(config.time_steps_decoder, config.num_decoder_tokens), name="decoder_inputs")
        decoder_lstm = LSTM(config.latent_dim, return_sequences=True, return_state=True, name='decoder_lstm')
        decoder_outputs, _, _ = decoder_lstm(decoder_inputs, initial_state=encoder_states)
        decoder_dense = Dense(config.num_decoder_tokens, activation='relu', name='decoder_relu')
        decoder_outputs = decoder_dense(decoder_outputs)

        # Combined model
        model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

        # Training data preprocessing
        training_list, validation_list = self.preprocessing()

        # Load dataset for training
        train = self.load_dataset(training_list)
        valid = self.load_dataset(validation_list)

        # Callbacks
        early_stopping = EarlyStopping(monitor='val_loss', patience=10, verbose=1, mode='min')
        opt = Adam(lr=0.0001)
        reduce_lr = keras.callbacks.ReduceLROnPlateau(monitor="val_loss",
                                                      factor=0.1, patience=50, verbose=0,
                                                      mode="auto")

        # Compile the model
        model.compile(metrics=['accuracy'], optimizer=opt, loss='categorical_crossentropy')

        validation_steps = len(validation_list) // self.batch_size
        steps_per_epoch = len(training_list) // self.batch_size

        # Add ModelCheckpoint callback to save the model with the highest validation accuracy
        checkpoint_filepath = os.path.join(self.save_model_path, 'best_model.h5')
        model_checkpoint = ModelCheckpoint(
            checkpoint_filepath,
            monitor='val_accuracy',
            save_best_only=True,
            mode='max',
            verbose=1
        )

        # Train the model
        model.fit(
            train,
            validation_data=valid,
            validation_steps=validation_steps,
            epochs=self.epochs,
            steps_per_epoch=steps_per_epoch,
            callbacks=[reduce_lr, model_checkpoint]  # Include the ModelCheckpoint callback
        )

        # Save the models
        if not os.path.exists(self.save_model_path):
            os.makedirs(self.save_model_path)

        self.encoder_model = Model(encoder_inputs, encoder_states)
        decoder_state_input_h = Input(shape=(self.latent_dim,))
        decoder_state_input_c = Input(shape=(self.latent_dim,))
        decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]
        decoder_outputs, state_h, state_c = decoder_lstm(
            decoder_inputs, initial_state=decoder_states_inputs)
        decoder_states = [state_h, state_c]
        decoder_outputs = decoder_dense(decoder_outputs)
        self.decoder_model = Model(
            [decoder_inputs] + decoder_states_inputs,
            [decoder_outputs] + decoder_states)
        # self.encoder_model.summary()
        # self.decoder_model.summary()

        # saving the models
        self.encoder_model.save(os.path.join(self.save_model_path, 'encoder_model.h5'))
        self.decoder_model.save_weights(os.path.join(self.save_model_path, 'decoder_model_weights.h5'))
        with open(os.path.join(self.save_model_path, 'tokenizer' + str(self.num_decoder_tokens)), 'wb') as file:
            joblib.dump(self.tokenizer, file)


if __name__ == "__main__":
    video_to_text = VideoDescriptionTrain(config)
    video_to_text.train_model()
```
